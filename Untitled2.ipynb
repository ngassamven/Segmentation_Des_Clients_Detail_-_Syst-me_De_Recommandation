  <p>RFM est une méthode utilisée pour analyser la valeur des clients et segmenter la base de clients. C'est un acronyme qui signifie :</p>
  <ul>
    <li><strong>Recency (R) :</strong> Ce métrique indique la date à laquelle un client a effectué son dernier achat. Une valeur de récence plus faible signifie que le client a acheté plus récemment, indiquant un engagement plus élevé avec la marque.</li>
    <li><strong>Frequency (F) :</strong> Ce métrique signifie la fréquence à laquelle un client effectue des achats dans une certaine période. Une valeur de fréquence plus élevée indique un client qui interagit plus souvent avec l'entreprise, suggérant une fidélité ou une satisfaction plus élevée.</li>
    <li><strong>Monetary (M) :</strong> Ce métrique représente le montant total d'argent qu'un client a dépensé sur une certaine période. Les clients ayant une valeur monétaire plus élevée ont contribué davantage à l'entreprise, indiquant leur potentiel de valeur à vie élevée.</li>
  </ul>
  <p>Ensemble, ces métriques aident à comprendre le comportement d'achat et les préférences d'un client, ce qui est essentiel pour personnaliser les stratégies marketing et créer un système de recommandation.</p>
</div>


<h3 style="color: black;">Étape 4.1.1 | Recency (R)</h3>

<div style="background-color: #b3cde0; color: black; border-radius: 15px; padding: 20px; width: 100%; max-width: 100%; box-sizing: border-box;">
  <h1 style="font-weight: bold;">Analyse de la Récence</h1>
  <p>Dans cette étape, nous nous concentrons sur la compréhension de la récence des achats effectués par un client. Il s'agit d'un aspect crucial de la segmentation des clients car il aide à identifier le niveau d'engagement des clients. Voici la fonctionnalité que nous allons définir :</p>
  <ul>
    <li><strong>Jours Depuis le Dernier Achat :</strong> Cette fonctionnalité représente le nombre de jours écoulés depuis le dernier achat du client. Une valeur plus faible indique que le client a acheté récemment, ce qui implique un niveau d'engagement plus élevé avec l'entreprise. En revanche, une valeur plus élevée peut indiquer une absence ou une diminution de l'engagement. En comprenant la récence des achats, les entreprises peuvent adapter leurs stratégies marketing pour réengager les clients qui n'ont pas effectué d'achats depuis un certain temps, ce qui peut potentiellement augmenter la rétention des clients et favoriser la fidélité.</li>
  </ul>
</div>


```python
# Convert InvoiceDate to datetime type
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Convert InvoiceDate to datetime and extract only the date
df['InvoiceDay'] = df['InvoiceDate'].dt.date

# Find the most recent purchase date for each customer
customer_data = df.groupby('CustomerID')['InvoiceDay'].max().reset_index()

# Find the most recent date in the entire dataset
most_recent_date = df['InvoiceDay'].max()

# Convert InvoiceDay to datetime type before subtraction
customer_data['InvoiceDay'] = pd.to_datetime(customer_data['InvoiceDay'])
most_recent_date = pd.to_datetime(most_recent_date)

# Calculate the number of days since the last purchase for each customer
customer_data['Days_Since_Last_Purchase'] = (most_recent_date - customer_data['InvoiceDay']).dt.days

# Remove the InvoiceDay column
customer_data.drop(columns=['InvoiceDay'], inplace=True)
```

<h3 style="color: black;">Étape 4.1.2 | Fréquence (F)</h3>

<div style="background-color: #b3cde0; color: black; border-radius: 15px; padding: 20px; width: 100%; max-width: 100%; box-sizing: border-box;">
  <h1 style="font-weight: bold;">Création des caractéristiques de fréquence d'engagement</h1>
  
  <p>Dans cette étape, je vais créer deux caractéristiques qui quantifient la fréquence de l'engagement d'un client avec le détaillant :</p>

  <h2 style="font-weight: bold;">Total Transactions :</h2>
  <p>Cette caractéristique représente le nombre total de transactions effectuées par un client. Elle aide à comprendre le niveau d'engagement d'un client avec le détaillant.</p>

  <h2 style="font-weight: bold;">Total Products Purchased :</h2>
  <p>Cette caractéristique indique le nombre total de produits (somme des quantités) achetés par un client au cours de toutes les transactions. Elle donne un aperçu du comportement d'achat du client en termes de volume de produits achetés.</p>
  
  <p>Ces caractéristiques seront cruciales pour segmenter les clients en fonction de leur fréquence d'achat, ce qui est un aspect clé pour déterminer les segments de clients pour un marketing ciblé et des recommandations personnalisées.</p>
</div>


```python
# Calculer le nombre total de transactions effectuées par chaque client
total_transactions = df.groupby('CustomerID')['InvoiceNo'].nunique().reset_index()
total_transactions.rename(columns={'InvoiceNo': 'Total_Transactions'}, inplace=True)

# Calculate the total number of products purchased by each customer
total_products_purchased = df.groupby('CustomerID')['Quantity'].sum().reset_index()
total_products_purchased.rename(columns={'Quantity': 'Total_Products_Purchased'}, inplace=True)

# Merge the new features into the customer_data dataframe
customer_data = pd.merge(customer_data, total_transactions, on='CustomerID')
customer_data = pd.merge(customer_data, total_products_purchased, on='CustomerID')

# Display the first few rows of the customer_data dataframe
customer_data.head()
```

<h3 style="color: black;">Étape 4.1.3 | Monétaire (M)</h3>

<div style="background-color: #b3cde0; color: black; border-radius: 15px; padding: 20px; width: 100%; max-width: 100%; box-sizing: border-box;">
  <h1 style="font-weight: bold;">Création des Caractéristiques Monétaires des Transactions des Clients</h1>
  
  <p>Dans cette étape, je vais créer deux caractéristiques qui représentent l'aspect monétaire des transactions des clients :</p>
  
  <h2 style="font-weight: bold;">Total Spend</h2>
  <p>Cette caractéristique représente le montant total dépensé par chaque client. Elle est calculée comme la somme du produit du prix unitaire (<em>UnitPrice</em>) et de la quantité (<em>Quantity</em>) pour toutes les transactions effectuées par un client. Cette caractéristique est cruciale car elle permet d'identifier le revenu total généré par chaque client, ce qui est un indicateur direct de la valeur d'un client pour l'entreprise.</p>
  
  <h2 style="font-weight: bold;">Average Transaction Value</h2>
  <p>Cette caractéristique est calculée comme le montant total dépensé divisé par le nombre total de transactions pour chaque client. Elle indique la valeur moyenne d'une transaction effectuée par un client. Ce critère est utile pour comprendre le comportement de dépense des clients par transaction, ce qui peut aider à adapter les stratégies de marketing et les offres à différents segments de clients en fonction de leurs habitudes de dépenses moyennes.</p>
</div>


```python
# Calculate the total spend by each customer
df['Total_Spend'] = df['UnitPrice'] * df['Quantity']
total_spend = df.groupby('CustomerID')['Total_Spend'].sum().reset_index()

# Calculate the average transaction value for each customer
average_transaction_value = total_spend.merge(total_transactions, on='CustomerID')
average_transaction_value['Average_Transaction_Value'] = average_transaction_value['Total_Spend'] / average_transaction_value['Total_Transactions']

# Merge the new features into the customer_data dataframe
customer_data = pd.merge(customer_data, total_spend, on='CustomerID')
customer_data = pd.merge(customer_data, average_transaction_value[['CustomerID', 'Average_Transaction_Value']], on='CustomerID')

# Display the first few rows of the customer_data dataframe
customer_data.head()
```

<h2 style="color: black;">Étape 4.6 | Saisonnière et Tendances</h2>

<div style="background-color: #b3cde0; color: black; border-radius: 15px; padding: 20px; width: 100%; max-width: 100%; box-sizing: border-box;">
  <h1 style="font-weight: bold;">Analyse de la Saison et des Tendances d'Achat :</h1>
  
  <p>Dans cette étape, je vais examiner la saisonnalité et les tendances dans les comportements d'achat des clients, ce qui peut offrir des informations précieuses pour adapter les stratégies marketing et améliorer la satisfaction client. Voici les caractéristiques que je souhaite introduire :</p>
  
  <h2 style="font-weight: bold;">Monthly_Spending_Mean :</h2>
  <p>Ceci est le montant moyen qu'un client dépense chaque mois. Cela nous aide à évaluer l'habitude générale de dépense de chaque client. Une moyenne plus élevée indique un client qui dépense plus, montrant potentiellement un intérêt pour les produits haut de gamme, tandis qu'une moyenne plus basse pourrait indiquer un client plus soucieux de son budget.</p>
  
  <h2 style="font-weight: bold;">Monthly_Spending_Std :</h2>
  <p>Cette caractéristique indique la variabilité des dépenses mensuelles d'un client. Une valeur plus élevée signale que les dépenses du client fluctuent considérablement d'un mois à l'autre, ce qui peut indiquer des achats importants sporadiques. En revanche, une valeur plus basse suggère des habitudes de dépenses plus stables et constantes. Comprendre cette variabilité peut aider à concevoir des promotions ou des réductions personnalisées pendant les périodes où les dépenses sont attendues pour être plus élevées.</p>
  
  <h2 style="font-weight: bold;">Spending_Trend :</h2>
  <p>Ceci reflète la tendance des dépenses d'un client au fil du temps, calculée comme la pente de la ligne de tendance linéaire ajustée aux données de dépenses. Une valeur positive indique une tendance à la hausse des dépenses, ce qui pourrait pointer vers une fidélité ou une satisfaction croissante. À l'inverse, une tendance négative pourrait signaler un intérêt ou une satisfaction en baisse, soulignant un besoin de stratégies de réengagement. Une valeur proche de zéro indique des habitudes de dépenses stables. Reconnaître ces tendances peut aider à développer des stratégies pour maintenir ou modifier les comportements de dépenses des clients, améliorant ainsi l'efficacité des campagnes marketing.</p>
  
  <p>En intégrant ces informations détaillées dans notre modèle de segmentation des clients, nous pouvons créer des groupes de clients plus précis et exploitables, facilitant le développement de stratégies marketing et de promotions hautement ciblées.</p>
</div>


```python
# Extract month and year from InvoiceDate
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month

# Calculate monthly spending for each customer
monthly_spending = df.groupby(['CustomerID', 'Year', 'Month'])['Total_Spend'].sum().reset_index()

# Calculate Seasonal Buying Patterns: We are using monthly frequency as a proxy for seasonal buying patterns
seasonal_buying_patterns = monthly_spending.groupby('CustomerID')['Total_Spend'].agg(['mean', 'std']).reset_index()
seasonal_buying_patterns.rename(columns={'mean': 'Monthly_Spending_Mean', 'std': 'Monthly_Spending_Std'}, inplace=True)

# Replace NaN values in Monthly_Spending_Std with 0, implying no variability for customers with single transaction month
seasonal_buying_patterns['Monthly_Spending_Std'].fillna(0, inplace=True)

# Calculate Trends in Spending 
# We are using the slope of the linear trend line fitted to the customer's spending over time as an indicator of spending trends
def calculate_trend(spend_data):
    # If there are more than one data points, we calculate the trend using linear regression
    if len(spend_data) > 1:
        x = np.arange(len(spend_data))
        slope, _, _, _, _ = linregress(x, spend_data)
        return slope
    # If there is only one data point, no trend can be calculated, hence we return 0
    else:
        return 0

# Apply the calculate_trend function to find the spending trend for each customer
spending_trends = monthly_spending.groupby('CustomerID')['Total_Spend'].apply(calculate_trend).reset_index()
spending_trends.rename(columns={'Total_Spend': 'Spending_Trend'}, inplace=True)

# Merge the new features into the customer_data dataframe
customer_data = pd.merge(customer_data, seasonal_buying_patterns, on='CustomerID')
customer_data = pd.merge(customer_data, spending_trends, on='CustomerID')

# Display the first few rows of the customer_data dataframe
customer_data.head()
```

  <h1 style="font-weight: bold;">Idetification et Traitement des valeurs manquantes  :</h1>
  
  <p>Dans cette section, je vais identifier et traiter les outliers dans notre ensemble de données. Les outliers sont des points de données qui diffèrent significativement de la majorité des autres points dans l'ensemble de données. Ces points peuvent potentiellement fausser les résultats de notre analyse, notamment dans le clustering K-means où ils peuvent influencer de manière significative la position des centres des clusters. Il est donc essentiel d'identifier et de traiter ces outliers de manière appropriée pour obtenir des résultats de clustering plus précis et significatifs.</p>
  
  <p>Étant donné la nature multidimensionnelle des données, il serait judicieux d'utiliser des algorithmes capables de détecter les outliers dans des espaces multidimensionnels. Je vais utiliser l'algorithme Isolation Forest pour cette tâche. Cet algorithme est efficace pour les données multidimensionnelles et est également computationalement efficace. Il isolement les observations en sélectionnant aléatoirement une caractéristique, puis en sélectionnant aléatoirement une valeur de découpage entre les valeurs maximales et minimales de la caractéristique sélectionnée.</p>
  
  <p>Passons à cette approche :</p>
</div>


```python
# Initializing the IsolationForest model with a contamination parameter of 0.05
model = IsolationForest(contamination=0.05, random_state=0)

# Fitting the model on our dataset (converting DataFrame to NumPy to avoid warning)
customer_data['Outlier_Scores'] = model.fit_predict(customer_data.iloc[:, 1:].to_numpy())

# Creating a new column to identify outliers (1 for inliers and -1 for outliers)
customer_data['Is_Outlier'] = [1 if x == -1 else 0 for x in customer_data['Outlier_Scores']]

# Display the first few rows of the customer_data dataframe
customer_data.head()
```

  <p>Il est essentiel de vérifier la corrélation entre les caractéristiques de notre ensemble de données. La présence de multicolinéarité, où les caractéristiques sont fortement corrélées, peut potentiellement affecter le processus de clustering en empêchant le modèle d'apprendre les véritables schémas sous-jacents dans les données, car les caractéristiques ne fournissent pas d'informations uniques. Cela pourrait conduire à des clusters qui ne sont pas bien séparés et significatifs.</p>
  
  <p>Si nous identifions de la multicolinéarité, nous pouvons utiliser des techniques de réduction de dimensionnalité comme la PCA (Analyse en Composantes Principales). Ces techniques aident à neutraliser l'effet de la multicolinéarité en transformant les caractéristiques corrélées en un nouvel ensemble de variables non corrélées, tout en préservant la majeure partie de la variance des données originales. Cette étape non seulement améliore la qualité des clusters formés, mais rend également le processus de clustering plus efficace sur le plan computationnel.</p>
</div>


```python
# Reset background style
sns.set_style('whitegrid')

# Calculate the correlation matrix excluding the 'CustomerID' column
corr = customer_data_cleaned.drop(columns=['CustomerID']).corr()

# Define a custom colormap
colors = ['#ff6200', '#ffcaa8', 'white', '#ffcaa8', '#ff6200']
my_cmap = LinearSegmentedColormap.from_list('custom_map', colors, N=256)

# Create a mask to only show the lower triangle of the matrix (since it's mirrored around its 
# top-left to bottom-right diagonal)
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask, k=1)] = True

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr, mask=mask, cmap=my_cmap, annot=True, center=0, fmt='.2f', linewidths=2)
plt.title('Correlation Matrix', fontsize=14)
plt.show()
```

<div style="background-color: #b3cde0; color: black; border-radius: 15px; padding: 20px; width: 100%; max-width: 100%; box-sizing: border-box;">
  <h1 style="font-weight: bold;">Préparation des Données :</h1>
  
  <p>Avant de passer au clustering et à la réduction de dimensions, il est impératif de normaliser nos caractéristiques. Cette étape est particulièrement importante, surtout dans le contexte des algorithmes basés sur les distances comme K-means et des méthodes de réduction de dimensions comme l'ACP. Voici pourquoi :</p>
  
  <ul>
    <li><strong>Pour le Clustering K-means :</strong> K-means repose fortement sur le concept de 'distance' entre les points de données pour former des clusters. Lorsque les caractéristiques ne sont pas sur une échelle similaire, les caractéristiques avec des valeurs plus grandes peuvent influencer de manière disproportionnée le résultat du clustering, conduisant potentiellement à des regroupements incorrects.</li>
    <li><strong>Pour l'ACP :</strong> L'ACP vise à trouver les directions où les données varient le plus. Lorsque les caractéristiques ne sont pas normalisées, celles avec des valeurs plus grandes peuvent dominer ces composants, ne reflétant pas fidèlement les motifs sous-jacents dans les données.</li>
  </ul>

  <h2 style="font-weight: bold;">Méthodologie :</h2>
  <p>Par conséquent, pour garantir une influence équilibrée sur le modèle et révéler les véritables motifs dans les données, je vais standardiser nos données, c'est-à-dire transformer les caractéristiques pour avoir une moyenne de 0 et un écart type de 1. Cependant, toutes les caractéristiques ne nécessitent pas de normalisation. Voici les exceptions et les raisons pour lesquelles elles sont exclues :</p>
  
  <ul>
    <li><strong>CustomerID :</strong> Cette caractéristique est simplement un identifiant pour les clients et ne contient aucune information significative pour le clustering.</li>
    <li><strong>Is_UK :</strong> Il s'agit d'une caractéristique binaire indiquant si le client est du Royaume-Uni ou non. Étant donné qu'elle prend déjà une valeur de 0 ou 1, la normaliser n'aura pas de différence significative.</li>
    <li><strong>Day_Of_Week :</strong> Cette caractéristique représente le jour de la semaine le plus fréquent auquel le client a effectué des transactions. Étant donné qu'il s'agit d'une caractéristique catégorique représentée par des entiers (1 à 7), la normaliser ne serait pas nécessaire.</li>
  </ul>
  
  <p>Je vais procéder à la normalisation des autres caractéristiques du jeu de données pour le préparer à l'ACP et au clustering K-means.</p>
</div>


```python
# Initialize the StandardScaler
scaler = StandardScaler()

# List of columns that don't need to be scaled
columns_to_exclude = ['CustomerID', 'Is_UK', 'Day_Of_Week']

# List of columns that need to be scaled
columns_to_scale = customer_data_cleaned.columns.difference(columns_to_exclude)

# Copy the cleaned dataset
customer_data_scaled = customer_data_cleaned.copy()

# Applying the scaler to the necessary columns in the dataset
customer_data_scaled[columns_to_scale] = scaler.fit_transform(customer_data_scaled[columns_to_scale])

# Display the first few rows of the scaled data
customer_data_scaled.head().style.background_gradient(cmap='Blues')
```

<div style="background-color: #b3cde0; color: black; border-radius: 15px; padding: 20px; width: 100%; max-width: 100%; box-sizing: border-box;">
  <h1 style="font-weight: bold;">Pourquoi avons-nous besoin de la réduction de dimensions ?</h1>
  
  <ul>
    <li><strong>Multicolinéarité Détectée :</strong> Dans les étapes précédentes, nous avons identifié que notre ensemble de données contient des caractéristiques multicolinéaires. La réduction de dimensions peut nous aider à éliminer les informations redondantes et à atténuer le problème de multicolinéarité.</li>
    <li><strong>Meilleur Clustering avec K-means :</strong> Étant donné que K-means est un algorithme basé sur la distance, un grand nombre de caractéristiques peut parfois diluer les motifs sous-jacents significatifs dans les données. En réduisant la dimensionnalité, nous pouvons aider K-means à trouver des clusters plus compacts et mieux séparés.</li>

  <p>Dans cette étape, nous envisageons l'application de techniques de réduction de dimensions pour simplifier nos données tout en conservant les informations essentielles. Parmi diverses méthodes telles que KernelPCA, ICA, ISOMAP, TSNE et UMAP, nous commençons par l'ACP (Analyse en Composantes Principales). Voici pourquoi :</p>
  
  <ul>
    <li><strong>ACP :</strong> L'ACP est un excellent point de départ car elle est efficace pour capturer les relations linéaires dans les données, ce qui est particulièrement pertinent compte tenu de la multicolinéarité identifiée dans notre ensemble de données. Elle nous permet de réduire le nombre de caractéristiques tout en conservant une part significative des informations, rendant ainsi notre analyse de clustering potentiellement plus précise et interprétable. De plus, elle est computationnellement efficace, ce qui signifie qu'elle n'augmentera pas considérablement le temps de traitement.</li>
  </ul>
  
  <p>Cependant, il est essentiel de noter que nous restons ouverts aux autres options. Après avoir appliqué l'ACP, si nous constatons que les premiers composants ne capturent pas une part significative de la variance, indiquant une perte d'informations vitales, nous pourrions envisager d'explorer d'autres méthodes non linéaires. Ces méthodes pourraient fournir une approche plus nuancée de la réduction de dimensions, capturant des motifs complexes que l'ACP pourrait manquer, bien que cela implique un temps de calcul et une complexité accrus.</p>
  
  <h2 style="font-weight: bold;">Méthodologie :</h2>
  
  <p>Je vais appliquer l'ACP sur tous les composants disponibles et tracer la variance cumulative expliquée par ces composants. Ce processus me permettra de visualiser combien de variance chaque composant principal supplémentaire peut expliquer, ce qui m'aidera à déterminer le nombre optimal de composants à retenir pour l'analyse.</p>
</div>


```python
# Setting CustomerID as the index column
customer_data_scaled.set_index('CustomerID', inplace=True)

# Apply PCA
pca = PCA().fit(customer_data_scaled)

# Calculate the Cumulative Sum of the Explained Variance
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

# Set the optimal k value (based on our analysis, we can choose 6)
optimal_k = 6

# Set seaborn plot style
sns.set(rc={'axes.facecolor': '#fcf0dc'}, style='darkgrid')

# Plot the cumulative explained variance against the number of components
plt.figure(figsize=(20, 10))

# Bar chart for the explained variance of each component
barplot = sns.barplot(x=list(range(1, len(cumulative_explained_variance) + 1)),
                      y=explained_variance_ratio,
                      color='#fcc36d',
                      alpha=0.8)

# Line plot for the cumulative explained variance
lineplot, = plt.plot(range(0, len(cumulative_explained_variance)), cumulative_explained_variance,
                     marker='o', linestyle='--', color='#ff6200', linewidth=2)

# Plot optimal k value line
optimal_k_line = plt.axvline(optimal_k - 1, color='red', linestyle='--', label=f'Optimal k value = {optimal_k}') 

# Set labels and title
plt.xlabel('Number of Components', fontsize=14)
plt.ylabel('Explained Variance', fontsize=14)
plt.title('Cumulative Variance vs. Number of Components', fontsize=18)

# Customize ticks and legend
plt.xticks(range(0, len(cumulative_explained_variance)))
plt.legend(handles=[barplot.patches[0], lineplot, optimal_k_line],
           labels=['Explained Variance of Each Component', 'Cumulative Explained Variance', f'Optimal k value = {optimal_k}'],
           loc=(0.62, 0.1),
           frameon=True,
           framealpha=1.0,  
           edgecolor='#ff6200')  

# Display the variance values for both graphs on the plots
x_offset = -0.3
y_offset = 0.01
for i, (ev_ratio, cum_ev_ratio) in enumerate(zip(explained_variance_ratio, cumulative_explained_variance)):
    plt.text(i, ev_ratio, f"{ev_ratio:.2f}", ha="center", va="bottom", fontsize=10)
    if i > 0:
        plt.text(i + x_offset, cum_ev_ratio + y_offset, f"{cum_ev_ratio:.2f}", ha="center", va="bottom", fontsize=10)

plt.grid(axis='both')   
plt.show()
```

<div style="background-color: #b3cde0; color: black; border-radius: 15px; padding: 20px; width: 100%; max-width: 100%; box-sizing: border-box;">
  <h1 style="font-weight: bold;">K-Means</h1>
  
  <p>K-Means est un algorithme d'apprentissage non supervisé qui regroupe les données en un nombre spécifié de groupes (K) en minimisant la somme des carrés intra-cluster (WCSS), également connue sous le nom d'inertie. L'algorithme attribue de manière itérative chaque point de données au centroïde le plus proche, puis met à jour les centroïdes en calculant la moyenne de tous les points attribués. Le processus se répète jusqu'à convergence ou jusqu'à ce qu'un critère d'arrêt soit atteint.</p>
  
  <h2 style="font-weight: bold;">Inconvénients de K-Means :</h2>
  <p><strong>1️⃣ Inertie influencée par le nombre de dimensions :</strong> La valeur de l'inertie tend à augmenter dans les espaces de haute dimension en raison de la malédiction de la dimensionnalité, ce qui peut déformer les distances euclidiennes entre les points de données.<br>
  <strong>Solution :</strong> Effectuer une réduction de dimensionnalité, telle que l'ACP, avant d'appliquer K-means pour atténuer ce problème et accélérer les calculs.</p>
  
  <p><strong>2️⃣ Dépendance à l'emplacement initial des centroïdes :</strong> L'algorithme K-means peut trouver un minimum local au lieu d'un minimum global, en fonction de l'endroit où les centroïdes sont initialement placés.<br>
  <strong>Solution :</strong> Pour améliorer la probabilité de localiser le minimum global, nous pouvons utiliser la méthode d'initialisation k-means++.</p>
  
  <p><strong>3️⃣ Exige la spécification du nombre de clusters :</strong> K-means nécessite la spécification du nombre de clusters (K) au préalable, ce qui peut ne pas être connu à l'avance.<br>
  <strong>Solution :</strong> Utiliser des méthodes telles que la méthode du coude et l'analyse de silhouette pour estimer le nombre optimal de clusters.</p>
  
  <p><strong>4️⃣ Sensibilité aux clusters de tailles inégales ou clairsemés :</strong> K-means peut avoir du mal avec des clusters de tailles ou de densités différentes.<br>
  <strong>Solution :</strong> Augmenter le nombre d'initialisations aléatoires (n_init) ou envisager d'utiliser des algorithmes qui gèrent mieux les clusters de tailles inégales, comme le GMM ou le DBSCAN.</p>
  
  <p><strong>5️⃣ Supposition de clusters convexes et isotropes :</strong> K-means suppose que les clusters sont sphériques et ont des variances similaires, ce qui n'est pas toujours le cas. Il peut avoir du mal avec des clusters allongés ou de forme irrégulière.<br>
  <strong>Solution :</strong> Envisager l'utilisation d'algorithmes de clustering qui ne font pas ces hypothèses, tels que DBSCAN ou le Modèle de Mélange Gaussien (GMM).</p>
  
  <p>En tenant compte des considérations ci-dessus, j'ai initialement appliqué l'ACP à l'ensemble de données. <h3 style="color: black;">Étape 9.1.1 | Méthode du Coude</h3>

<div style="background-color: #b3cde0; color: black; border-radius: 15px; padding: 20px; width: 100%; max-width: 100%; box-sizing: border-box;">
  <h1 style="font-weight: bold;">Méthode du Coude</h1>
  
  <p>La Méthode du Coude est une technique pour identifier le nombre idéal de clusters dans un ensemble de données. Elle consiste à itérer à travers les données en générant des clusters pour différentes valeurs de <strong>k</strong>. L'algorithme K-means calcule la somme des distances au carré entre chaque point de données et son centroïde de cluster assigné, connue sous le nom de score d'inertie ou WCSS (Within-Cluster Sum of Squares). En traçant le score d'inertie en fonction de la valeur de <strong>k</strong>, nous créons un graphique qui présente généralement une forme de coude, d'où le nom "Méthode du Coude". Le point du coude représente la valeur de <strong>k</strong> où la réduction de l'inertie obtenue en augmentant <strong>k</strong> devient négligeable, indiquant le point d'arrêt optimal pour le nombre de clusters.</p>
  
  <h2 style="font-weight: bold;">Utilisation de la Bibliothèque YellowBrick</h2>
  
  <p>Dans cette section, je vais utiliser la bibliothèque YellowBrick pour faciliter la mise en œuvre de la Méthode du Coude. YellowBrick, une extension de l'API Scikit-Learn, est réputée pour sa capacité à générer rapidement des visualisations percutantes dans le domaine de l'apprentissage automatique.</p>

<h2 style="color: black;">Étape 9.1.2 | Méthode de Silhouette</h2>

<div style="background-color: #b3cde0; color: black; border-radius: 15px; padding: 20px; width: 100%; max-width: 100%; box-sizing: border-box;">
  <h1 style="font-weight: bold;">Méthode de Silhouette</h1>
  
  <p>La Méthode de Silhouette est une approche pour trouver le nombre optimal de clusters dans un ensemble de données en évaluant la cohérence au sein des clusters et leur séparation par rapport aux autres clusters. Elle calcule le coefficient de silhouette pour chaque point de données, ce qui mesure à quel point un point est similaire à son propre cluster par rapport aux autres clusters.</p>
  
  <h2 style="font-weight: bold;">Qu'est-ce que le Coefficient de Silhouette ?</h2>
  
  <p>Pour déterminer le coefficient de silhouette pour un point donné <strong>i</strong>, suivez ces étapes :</p>
  
  <ol>
    <li><strong>Calculer a(i) :</strong> Calculez la distance moyenne entre le point <strong>i</strong> et tous les autres points de son cluster.</li>
    <li><strong>Calculer b(i) :</strong> Calculez la distance moyenne entre le point <strong>i</strong> et tous les points du cluster le plus proche de son propre cluster.</li>
    <li><strong>Calculer le coefficient de silhouette, s(i), pour le point <strong>i</strong> en utilisant la formule suivante :</strong></li>
  </ol>
  
  <p>
    <strong>s(i) = (b(i) - a(i)) / max(b(i), a(i))</strong>
  </p>
  
  <p>Note : Le coefficient de silhouette quantifie la similarité d'un point avec son propre cluster (cohésion) par rapport à sa séparation avec les autres clusters. Cette valeur varie de -1 à 1, avec des valeurs plus élevées signifiant que le point est bien aligné avec son cluster et a une faible similarité avec les clusters voisins.</p>
  
  <h2 style="font-weight: bold;">Qu'est-ce que le Score de Silhouette ?</h2>
  
  <p>Le score de silhouette est le coefficient de silhouette moyen calculé pour tous les points de données dans un ensemble de données. Il fournit une évaluation globale de la qualité du clustering, prenant en compte à la fois la cohésion au sein des clusters et la séparation entre les clusters. Un score de silhouette plus élevé indique une meilleure configuration de clustering.</p>
  
  <h2 style="font-weight: bold;">Quels sont les Avantages de la Méthode de Silhouette par rapport à la Méthode du Coude ?</h2>
  
  <ul>
    <li><strong>Évaluation de la qualité des clusters :</strong> La Méthode de Silhouette évalue la qualité des clusters en considérant à la fois la cohésion au sein des clusters et leur séparation par rapport aux autres clusters. Cela fournit une mesure plus complète de la performance du clustering par rapport à la Méthode du Coude, qui ne considère que l'inertie (somme des distances au carré au sein des clusters).</li>
    <li><strong>Quantification directe :</strong> La Méthode de Silhouette produit un score de silhouette qui quantifie directement la qualité du clustering, facilitant la comparaison de différentes valeurs de <strong>k</strong>. En revanche, la Méthode du Coude repose sur l'interprétation subjective du point de coude, ce qui peut être moins fiable lorsque le graphique ne montre pas un coude clair.</li>
    <li><strong>Visualisation des coefficients :</strong> La Méthode de Silhouette génère une représentation visuelle des coefficients de silhouette pour chaque point de données, permettant une identification plus facile des fluctuations et des points atypiques au sein des clusters. Cela aide à déterminer le nombre optimal de clusters avec une plus grande confiance, contrairement à la Méthode du Coude, qui repose sur l'inspection visuelle du graphique de l'inertie.</li>
  </ul>
</div>

    <li><strong>Choix du nombre de clusters :</strong> Je commencerai par choisir une gamme de valeurs pour <strong>k</strong> allant de 2 à 6, basée sur la Méthode du Coude de la section précédente. Ensuite, je tracerai les scores de Silhouette pour chaque valeur de <strong>k</strong> afin de déterminer celle avec le score le plus élevé.</li>
    <li><strong>Ajustement du choix du <strong>k</strong> :</strong> Par la suite, pour affiner la sélection du <strong>k</strong> le plus approprié, je générerai des graphiques de Silhouette qui affichent visuellement les coefficients de silhouette pour chaque point de données au sein des différents clusters.</li>
    <li><strong>Utilisation de YellowBrick :</strong> La bibliothèque YellowBrick sera de nouveau utilisée pour créer ces graphiques et faciliter une analyse comparative.</li>
  </ol>

```python
def silhouette_analysis(df, start_k, stop_k, figsize=(15, 16)):
    """
    Perform Silhouette analysis for a range of k values and visualize the results.
    """

    # Set the size of the figure
    plt.figure(figsize=figsize)

    # Create a grid with (stop_k - start_k + 1) rows and 2 columns
    grid = gridspec.GridSpec(stop_k - start_k + 1, 2)

    # Assign the first plot to the first row and both columns
    first_plot = plt.subplot(grid[0, :])

    # First plot: Silhouette scores for different k values
    sns.set_palette(['darkorange'])

    silhouette_scores = []

    # Iterate through the range of k values
    for k in range(start_k, stop_k + 1):
        km = KMeans(n_clusters=k, init='k-means++', n_init=10, max_iter=100, random_state=0)
        km.fit(df)
        labels = km.predict(df)
        score = silhouette_score(df, labels)
        silhouette_scores.append(score)

    best_k = start_k + silhouette_scores.index(max(silhouette_scores))

    plt.plot(range(start_k, stop_k + 1), silhouette_scores, marker='o')
    plt.xticks(range(start_k, stop_k + 1))
    plt.xlabel('Number of clusters (k)')
    plt.ylabel('Silhouette score')
    plt.title('Average Silhouette Score for Different k Values', fontsize=15)

    # Add the optimal k value text to the plot
    optimal_k_text = f'The k value with the highest Silhouette score is: {best_k}'
    plt.text(10, 0.23, optimal_k_text, fontsize=12, verticalalignment='bottom', 
             horizontalalignment='left', bbox=dict(facecolor='#fcc36d', edgecolor='#ff6200', boxstyle='round, pad=0.5'))
             

    # Second plot (subplot): Silhouette plots for each k value
    colors = sns.color_palette("bright")

    for i in range(start_k, stop_k + 1):    
        km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=100, random_state=0)
        row_idx, col_idx = divmod(i - start_k, 2)

        # Assign the plots to the second, third, and fourth rows
        ax = plt.subplot(grid[row_idx + 1, col_idx])

        visualizer = SilhouetteVisualizer(km, colors=colors, ax=ax)
        visualizer.fit(df)

        # Add the Silhouette score text to the plot
        score = silhouette_score(df, km.labels_)
        ax.text(0.97, 0.02, f'Silhouette Score: {score:.2f}', fontsize=12, \
                ha='right', transform=ax.transAxes, color='red')

        ax.set_title(f'Silhouette Plot for {i} Clusters', fontsize=15)

    plt.tight_layout()
    plt.show()
```

<h2 style="color: black;">Étape 9.2 | Modèle de Clustering - K-meanss</h2>

<div style="background-color: #e0f7fa; color: black; border-radius: 15px; padding: 20px; width: 100%; max-width: 100%; box-sizing: border-box;">
  <h1 style="font-weight: bold;">Application de l'Algorithme K-means</h1>
  
  <p>Dans cette étape, je vais appliquer l'algorithme de clustering K-means pour segmenter les clients en différents groupes en fonction de leurs comportements d'achat et d'autres caractéristiques, en utilisant le nombre optimal de clusters déterminé à l'étape précédente.</p>
  
  <p>Il est important de noter que l'algorithme K-means peut attribuer des étiquettes différentes aux clusters à chaque exécution. Pour remédier à cela, nous avons pris une mesure supplémentaire pour échanger les étiquettes en fonction de la fréquence des échantillons dans chaque cluster, garantissant ainsi une attribution cohérente des étiquettes à travers différentes exécutions</p>
/div>
s.

```python
# Apply KMeans clustering using the optimal k
kmeans = KMeans(n_clusters=3, init='k-means++', n_init=10, max_iter=100, random_state=0)
kmeans.fit(customer_data_pca)

# Get the frequency of each cluster
cluster_frequencies = Counter(kmeans.labels_)

# Create a mapping from old labels to new labels based on frequency
label_mapping = {label: new_label for new_label, (label, _) in 
                 enumerate(cluster_frequencies.most_common())}

# Reverse the mapping to assign labels as per your criteria
label_mapping = {v: k for k, v in {2: 1, 1: 0, 0: 2}.items()}

# Apply the mapping to get the new labels
new_labels = np.array([label_mapping[label] for label in kmeans.labels_])

# Append the new cluster labels back to the original dataset
customer_data_cleaned['cluster'] = new_labels

# Append the new cluster labels to the PCA version of the dataset
customer_data_pca['cluster'] = new_labels
```

```python
# Setting 'CustomerID' column as index and assigning it to a new dataframe
df_customer = customer_data_cleaned.set_index('CustomerID')

# Standardize the data (excluding the cluster column)
scaler = StandardScaler()
df_customer_standardized = scaler.fit_transform(df_customer.drop(columns=['cluster'], axis=1))

# Create a new dataframe with standardized values and add the cluster column back
df_customer_standardized = pd.DataFrame(df_customer_standardized, columns=df_customer.columns[:-1], index=df_customer.index)
df_customer_standardized['cluster'] = df_customer['cluster']

# Calculate the centroids of each cluster
cluster_centroids = df_customer_standardized.groupby('cluster').mean()

# Function to create a radar chart
def create_radar_chart(ax, angles, data, color, cluster):
    # Plot the data and fill the area
    ax.fill(angles, data, color=color, alpha=0.4)
    ax.plot(angles, data, color=color, linewidth=2, linestyle='solid')
    
    # Add a title
    ax.set_title(f'Cluster {cluster}', size=20, color=color, y=1.1)

# Set data
labels=np.array(cluster_centroids.columns)
num_vars = len(labels)

# Compute angle of each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# The plot is circular, so we need to "complete the loop" and append the start to the end
labels = np.concatenate((labels, [labels[0]]))
angles += angles[:1]

# Initialize the figure
fig, ax = plt.subplots(figsize=(20, 10), subplot_kw=dict(polar=True), nrows=1, ncols=3)

# Create radar chart for each cluster
for i, color in enumerate(colors):
    data = cluster_centroids.loc[i].tolist()
    data += data[:1]  # Complete the loop
    create_radar_chart(ax[i], angles, data, color, i)

# Add input data
ax[0].set_xticks(angles[:-1])
ax[0].set_xticklabels(labels[:-1])

ax[1].set_xticks(angles[:-1])
ax[1].set_xticklabels(labels[:-1])

ax[2].set_xticks(angles[:-1])
ax[2].set_xticklabels(labels[:-1])

# Add a grid
ax[0].grid(color='grey', linewidth=0.5)

# Display the plot
plt.tight_layout()
plt.show()
```

  <h1 style="font-weight: bold;">Interprétation</h1>

  <p>Dans la phase finale de ce projet, je vais développer un système de recommandation pour améliorer l’expérience d’achat en ligne. Ce système suggérera des produits aux clients en fonction des habitudes d’achat prédominantes dans leurs clusters respectifs.

Auparavant, lors de la préparation des données clients, j’ai isolé une petite fraction (5%) des clients identifiés comme des anomalies et les ai réservés dans un ensemble de données distinct appelé outliers_data.

Maintenant, en me concentrant sur les 95% principaux du groupe de clients, j’analyse les données clients nettoyées pour identifier les produits les plus vendus dans chaque cluster. En utilisant ces informations, le système créera des recommandations personnalisées, suggérant les trois produits les plus populaires au sein de leur cluster qu’ils n’ont pas encore achetés. Cela facilite non seulement les stratégies de marketing ciblées, mais enrichit également l’expérience d’achat personnelle, augmentant potentiellement les ventes. Pour le groupe des anomalies, une approche de base pourrait consister à recommander des produits aléatoires, comme point de départ pour les engager.</p>

```python
# Step 1: Extract the CustomerIDs of the outliers and remove their transactions from the main dataframe
outlier_customer_ids = outliers_data['CustomerID'].astype('float').unique()
df_filtered = df[~df['CustomerID'].isin(outlier_customer_ids)]

# Step 2: Ensure consistent data type for CustomerID across both dataframes before merging
customer_data_cleaned['CustomerID'] = customer_data_cleaned['CustomerID'].astype('float')

# Step 3: Merge the transaction data with the customer data to get the cluster information for each transaction
merged_data = df_filtered.merge(customer_data_cleaned[['CustomerID', 'cluster']], on='CustomerID', how='inner')

# Step 4: Identify the top 10 best-selling products in each cluster based on the total quantity sold
best_selling_products = merged_data.groupby(['cluster', 'StockCode', 'Description'])['Quantity'].sum().reset_index()
best_selling_products = best_selling_products.sort_values(by=['cluster', 'Quantity'], ascending=[True, False])
top_products_per_cluster = best_selling_products.groupby('cluster').head(10)

# Step 5: Create a record of products purchased by each customer in each cluster
customer_purchases = merged_data.groupby(['CustomerID', 'cluster', 'StockCode'])['Quantity'].sum().reset_index()

# Step 6: Generate recommendations for each customer in each cluster
recommendations = []
for cluster in top_products_per_cluster['cluster'].unique():
    top_products = top_products_per_cluster[top_products_per_cluster['cluster'] == cluster]
    customers_in_cluster = customer_data_cleaned[customer_data_cleaned['cluster'] == cluster]['CustomerID']
    
    for customer in customers_in_cluster:
        # Identify products already purchased by the customer
        customer_purchased_products = customer_purchases[(customer_purchases['CustomerID'] == customer) & 
                                                         (customer_purchases['cluster'] == cluster)]['StockCode'].tolist()
        
        # Find top 3 products in the best-selling list that the customer hasn't purchased yet
        top_products_not_purchased = top_products[~top_products['StockCode'].isin(customer_purchased_products)]
        top_3_products_not_purchased = top_products_not_purchased.head(3)
        
        # Append the recommendations to the list
        recommendations.append([customer, cluster] + top_3_products_not_purchased[['StockCode', 'Description']].values.flatten().tolist())

# Step 7: Create a dataframe from the recommendations list and merge it with the original customer data
recommendations_df = pd.DataFrame(recommendations, columns=['CustomerID', 'cluster', 'Rec1_StockCode', 'Rec1_Description', \
                                                 'Rec2_StockCode', 'Rec2_Description', 'Rec3_StockCode', 'Rec3_Description'])
customer_data_with_recommendations = customer_data_cleaned.merge(recommendations_df, on=['CustomerID', 'cluster'], how='right')
```
